```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
# Create 2 bionomial populations with n=10, size= 10000, p =0.2 or 0.5
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 
```

```python
# Plot the population
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
# Create 2 samples from populations
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
```

## 1.1 Increase sample size from 100 to 1000
We should expect the means stay relative the same but the standard deviations decrease. 

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
```

## 1.2 Decrease Sample size 20 from the population

```python
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
```

With more sample size (100, or 1000), the mean of the sample would be closer to the mean (n*p = 2 or 5) of the population.  When sample size is small (20), the mean can change a little but standard deviation does not change much.  In large sample size, the mean and std are a good presentative to describe the central tendancy of the data. 

# 2. Change p to 0.3 in pop1

```python
# Change p=0.3 in population 1. 
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 
# Take 100 samples from each poluation
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

#find out sample mean and std
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
print (ttest_ind(sample2, sample1, equal_var=False))

```

The p value < 0.01 indicate that there is significant difference between sample 1 and sample 2.

```python
# Change p=0.3 in population 1. 
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 
# Take 100 samples from each poluation
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

#find out sample mean and std
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
print (ttest_ind(sample2, sample1, equal_var=False))
```

The p value < 0.01 indicate that there is significant difference between sample 1 and sample 2.  However, T-statistic value decreases and p value increases, suggesting that the difference between the two samples are smaller.  

```python
# use a normal distribution
pop1 = np.random.normal(2, 1, 10000)
pop2 = np.random.normal(5, 0.5, 10000) 
# Take 100 samples from each poluation
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

#find out sample mean and std
print (sample1.mean())
print (sample1.std())
print (sample2.mean())
print (sample2.std())
print (ttest_ind(sample2, sample1, equal_var=False))
```
